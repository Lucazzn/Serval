```python
import pickle
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns
import sys
import os
sys.path.append('/home/yutao4/Sat_Simulator')
os.chdir('/home/yutao4/Sat_Simulator')
```

```python
sns.set_theme()
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['font.size'] = 18
```

```python
weather_record=pickle.load(open("results/weather.pkl", "rb"))
plt.figure(figsize=(6, 4))
name_mapping={
    'no_weather.log (high priority)': 'Serval (no weather)',
    'weather.log (high priority)': 'Serval',
}
for name in weather_record:
    plt.plot(weather_record[name][0], weather_record[name][1], label=f"{name_mapping[name]}")
plt.legend()
plt.xlabel("Latency (hours)")
plt.ylabel("CDF")
plt.xscale("log")
plt.savefig("results/plots/weather.pdf", dpi=300, bbox_inches='tight')
plt.show()
```

```python
compute_record=pickle.load(open("results/compute.pkl", "rb"))
name_mapping={
    'stop_compute.log (high priority)': 'Serval (no compute)',
    'default.log (high priority)': 'Serval',
    # 'Serval_fast_compute (high priority)': 'Serval (fast compute)',
}
plt.figure(figsize=(6, 4))
for name in compute_record:
    if "fast_compute" in name:
        continue
    plt.plot(compute_record[name][0], compute_record[name][1], label=f"{name_mapping[name]}")
plt.legend()
plt.xlabel("Latency (hours)")
plt.ylabel("CDF")
plt.xscale("log")
plt.savefig("results/plots/compute.pdf", dpi=300, bbox_inches='tight')
```

```python
forest_record=pickle.load(open("results/forest_old.pkl", "rb"))
name_mapping={
    'default_filter': 'Serval (no forest)',
    'forest_model': 'Serval',
}
plt.figure(figsize=(6, 4))
for name in forest_record:
    short_name = name.split("__")[0]
    if short_name not in name_mapping:
        continue
    plt.plot(forest_record[name][0], forest_record[name][1], label=f"{name_mapping[short_name]}")
plt.legend()
plt.xlabel("Latency (hours)")
plt.ylabel("CDF")
plt.xscale("log")
plt.savefig("results/plots/forest.pdf", dpi=300, bbox_inches='tight')
```

```python
prior_sweep_record=pickle.load(open("results/priority_sweep.pkl", "rb"))
plt.figure(figsize=(6, 4))

colors = {
    'Serval 0.5%': (0.2980392156862745, 0.4470588235294118, 0.6901960784313725),
    'Serval 0.1%': (0.7686274509803922, 0.3058823529411765, 0.3215686274509804),
    'Serval 0.05%': (0.8666666666666667, 0.5176470588235295, 0.3215686274509804),
    'In-Order': (0.3333333333333333, 0.6588235294117647, 0.40784313725490196)
}
#lst.sort()

for name in colors:
    short_name = name.split("__")[0]
    l = plt.plot(prior_sweep_record[name][0], prior_sweep_record[name][1], label=f"{short_name}", color=colors[name])

plt.legend()
plt.xlabel("Latency (hours)")
plt.ylabel("CDF")
plt.xscale("log")
plt.savefig("results/plots/priority_sweep.pdf", dpi=300, bbox_inches='tight')
```
